```python
import numpy as np
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from research.logreg import fit_and_score, pred_grid, show_pred
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
X, y = make_moons(n_samples=1000, noise=0.17, random_state=14)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=13)
```

```python
train_test = (X_train, X_test, y_train, y_test)
results = {degree: fit_and_score(degree=degree, train_test=train_test, C=1.0) for degree in range(1, 10)}
for result in sorted(results):
    print(f'{f" Model results for degree: {result} ":*^80}\n')
    print(results[result][1])
```

```python
lrbt = -2.0, 3.0, -1.5, 2.0
n = 100
show_pred(results[3][0], X_train, y_train, lrbt, n)
```
